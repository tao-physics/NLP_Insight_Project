```python
(data_pred[data_pred['Label_encoded'] == data_pred['label_2ndpred']]['prob_2nd'] > 0.1).mean()
```

```python
label_final_pred = []
for k in range(len(data_pred)):
    temp = data_pred['label_pred'][k]
    if data_pred['Label_encoded'][k] == data_pred['label_2ndpred'][k]:
        temp = data_pred['label_2ndpred'][k]
    label_final_pred.append(temp)
    
data_pred['label_final_pred'] = label_final_pred
```

```python
data_pred['Label_encoded'][0]
```

```python
data_pred.head()
```

```python
print(classification_report(data_pred['Label_encoded'], data_pred['label_final_pred']))
```
